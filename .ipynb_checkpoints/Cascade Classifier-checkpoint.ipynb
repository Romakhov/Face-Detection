```python
import cv2
import matplotlib.pyplot as plt
import dlib
from imutils import face_utils
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
faceCascade = cv2.CascadeClassifier("CascadeClassifiers/haarcascade_frontalface_default.xml")
eyeCascade = cv2.CascadeClassifier("CascadeClassifiers/haarcascade_eye.xml")
smileCascade = cv2.CascadeClassifier("CascadeClassifiers/haarcascade_smile.xml")
```

```python
# Load the image
image = cv2.imread('images/img.jpg')
plt.figure(figsize=(20,8))
plt.imshow(image, cmap='gray')
plt.show()
```

```python
# Detect faces
faces = faceCascade.detectMultiScale(
image,
scaleFactor=1.1,
minNeighbors=5,
flags=cv2.CASCADE_SCALE_IMAGE
)

eyes = eyeCascade.detectMultiScale(
image,
scaleFactor=1.1,
minNeighbors=5,
flags=cv2.CASCADE_SCALE_IMAGE
)
# For each face
for (x, y, w, h) in faces: 
    # Draw rectangle around the face
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 255, 255), 3)
    
for (x, y, w, h) in eyes: 
    # Draw rectangle around the face
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 255, 255), 3)    
```

```python
plt.figure(figsize=(12,8))
plt.imshow(image, cmap='gray')
plt.show()
```
